```python
from tensorflow.keras.layers import Conv2D, MaxPooling2D
```

```python
m3_input = Input(shape=(150,150,3))
m3_conv = Conv2D(50, 3, activation = 'relu')(m3_input)
m3_mp = MaxPooling2D(2,2)(m3_conv)
m3_f = Flatten()(m3_mp)
m3_x = Dense(100, activation = 'relu')(m3_f)
m3_outputs = Dense(10, activation = 'softmax')(m3_x)

model_3 = Model(inputs = m3_input, outputs = m3_outputs)

model_3.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['acc'])


```

```python
model_3.fit(train_generator, epochs=200, validation_data= test_generator, verbose=1)
```

```python
from tensorflow.keras.applications.resnet import ResNet101
```

```python
resnet_model = ResNet101(include_top = False, weights = 'imagenet', input_shape = (150,150,3))
```

```python
for layer in resnet_model.layers:
    layer.trainable = False
```

```python
resnet_model.summary()
```

```python
for layer in resnet_model.layers[-20:]:
    print(layer.name)
```

```python
from tensorflow.keras.layers import Dropout
```

```python
last_layer = resnet_model.get_layer('conv5_block3_out').output

x1 = Flatten()(last_layer)
x2 = Dense(1024, activation = 'relu')(x1)
x3 = Dropout(0.3)(x2)
output_resnet = Dense(10, activation = 'softmax')(x3)

final_resnet_model = Model(inputs = resnet_model.input, outputs = output_resnet)

final_resnet_model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['acc'])

```

```python
final_resnet_model.fit(train_generator, epochs=200, validation_data=test_generator, verbose=1)
```
